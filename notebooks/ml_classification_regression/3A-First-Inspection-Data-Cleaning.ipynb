```python
# External Dependencies
import numpy as np
import pandas as pd
```

```python
# Read in pickled combined data
X_y_data = pd.read_pickle(relative_filepath + "data/interim/step_2/X_y_data.pkl")

# Read in pickled train data
X_y_train = pd.read_pickle(relative_filepath + "data/interim/step_2/X_y_train.pkl")

# Read in pickled test data
X_y_test = pd.read_pickle(relative_filepath + "data/interim/step_2/X_y_test.pkl")

# Recap data structure of combined
X_y_data.head()
X_y_data.shape
```

```python
#values for config dict
input_dfs = [X_y_data,
             X_y_train,
             X_y_test]

target = "classLabel"
```

## d) Initial Data Preprocessing

#### i) Encode target variable

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()
le.fit(X_y_train[target])

list(le.classes_)
list(le.inverse_transform([0, 1]))

X_y_data[target] = le.transform(X_y_data[target])
X_y_train[target] = le.transform(X_y_train[target])
X_y_test[target] = le.transform(X_y_test[target])
```

### iii) Actions from first inspection

```python
# Convert comma floats to point floats
conversion_cols = ['v2', 'v3', 'v5', 'v6', 'v7', 'v15']

for df in input_dfs:
    for col in conversion_cols:
        
         df[col] = df[col].apply(lambda x: pd.to_numeric(str(x).replace(',','.'), errors='coerce'))   
    
    df.head()
    df.shape
```
